## Load the Dataset

```python
import pandas as pd

df = pd.read_csv('Google_Stock_Price_Train.csv')
df['Open'] = df['Open'].astype(str).str.replace(',', '').astype(float)  # Force to str before replace
df = df[['Date', 'Open']]  # Keep only relevant columns
```

```python
import matplotlib.pyplot as plt

plt.plot(df['Open'])
plt.title("Google Stock Opening Price Over Time")
plt.xlabel("Days")
plt.ylabel("Open Price")
plt.show()
```

```python
# Step 1: Load the test dataset
test_df = pd.read_csv('Google_Stock_Price_Test.csv')

# Step 2: Clean the 'Open' column (same as training data)
test_df['Open'] = test_df['Open'].astype(str).str.replace(',', '').astype(float)

# Optional: Keep only 'Open' if you're using only that
real_stock_price = test_df['Open'].values  # This will be used for comparison later
```

## Feature Scaling (Normalization) 
### LSTM works better when input values are between 0 and 1.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df[['Open']])
```

## Prepare Inputs for Prediction Using Last 60 Days + Test Set
### LSTM expects sequences. We use 60 previous days to predict the next day's price.

```python
# Use the last 60 days of the training data + test data to create prediction input
total_data = pd.concat((df['Open'], test_df['Open']), axis=0)

# Only the part we need: 60 before test + test length
model_input = total_data[len(total_data) - len(test_df) - 60:].values.reshape(-1, 1)

# Apply the same scaling
model_input_scaled = scaler.transform(pd.DataFrame(model_input, columns=['Open']))

# Create X_test sequences
X_test = []
for i in range(60, 60 + len(test_df)):
    X_test.append(model_input_scaled[i-60:i, 0])

import numpy as np
X_test = np.array(X_test)
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

##  Build the LSTM Model
### LSTM remembers past values in sequences.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(60, 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))  # Output layer for predicting a single value
```
